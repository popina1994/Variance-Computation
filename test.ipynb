```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
std_dev = 1000
mean = 100
num_elems = 1000000
array  = mean + std_dev * np.random.standard_normal(size=num_elems)

#plt.plot(array, 'bo')
plt.hist(array, 100, density=True, facecolor="blue")
```

```python
sum = 0.0
for num in array:
    #print(num)
    sum += num

print(sum)
print("Computed mean {:.15f}".format(sum / num_elems))
print(sum / num_elems)
```
